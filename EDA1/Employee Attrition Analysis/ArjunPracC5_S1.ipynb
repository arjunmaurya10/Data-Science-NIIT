**Task2**

```python
def data_types(df):
    cat_var=[]
    con_var=[]
    
    
    for col in df:
        if df[col].nunique()<43:
            cat_var.append(col)
        else:
            con_var.append(col)
    return cat_var,con_var       
```

```python
categorial_var, continous_var = data_types(emp_df)
print(tabulate({"cat_var":categorial_var,"con_var":continous_var},headers=["Categorial Var", "Continous Var"]))
```

**Task3**

```python
def cat_info(col):
    print(f"Unique values in {col} : {emp_df[col].unique()}")
    print(f"Missing Value counts of {col}: {emp_df[col].isnull().sum()}")
    print(f"Mode of {col}: {emp_df[col].mode()[0]}")
    value_counts=emp_df[col].value_counts()
    print("value counts:", value_counts)
    
```

```python
def cat_visual(col):
    value_counts= emp_df[col].value_counts()
    per= emp_df[col].value_counts(normalize=True)*100 #value_counts in percentage
    plt.figure(figsize=(8,6))
    plt.bar(value_counts.index,value_counts.values)
    plt.title(f"value count of{col}")
        
    
    plt.figure(figsize=(8,6))
    plt.pie(per.values,labels=per.index,autopct="%1.1f%%")
    plt.title(f"percentage of {col}")
```

```python
for col in categorial_var:
    cat_visual(col)
    print("\n")
```

The above bar graphs and pie charts shows the count and distribution of percentagea for each of the Categorical Values . We can see that Maximum no. pf employees are of **age 34 and 35**.

```python
def num_info(col):
    summary=emp_df[col].describe()
    print(summary)
    print(f"median of {col}: {emp_df[col].median()}")
    print(f"missing values {col}: {emp_df[col].isnull().sum()}")
    print(f"skewness of {col}:{emp_df[col].skew()}")
    print(f"Kurtosis of {col}: {emp_df[col].kurt()}")
```

```python
def num_visual(col):
    fig,ax=plt.subplots(2,1,figsize=(10,10))
    ax[0].hist(emp_df[col])
    sns.boxplot(y=emp_df[col],ax=ax[1])
```

Two functions have been created to calculate the numerical data for the Continous Variables and a function to represent them in a chart or graph was created called num_info and num_visual repectively.

**Outliers Detected**

```python
mti_16=emp_df[emp_df["MonthlyIncome"]>=11000].index
for i in mti_16:
    emp_df.loc[i,"MonthlyIncome"]=emp_df["MonthlyIncome"].mean()
```

```python
num_visual('MonthlyIncome')
```

**Outliers treated**

```python
sns.boxplot(y=emp_df["MonthlyIncome"])
```

From the above calculations outliers were found and were treated by calculating the mean and after the process the boxplot was created without any Outliers .
