# DBSCAN 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(42)
def PointsInCircum(r,n=100):
    return[(math.cos(2*math.pi/n*x)*r + np.random.normal(-20,20),
            math.sin(2*math.pi/n*x)*r + np.random.normal(-20,20)) for x in range(1,n+1)]
```

```python
a = PointsInCircum(400,900)
```

```python
a
```

```python
b= PointsInCircum(250,600)
b
```

```python
c=PointsInCircum(90,200)
c
```

```python
a = pd.DataFrame(a)
```

```python
a
```

```python
b = pd.DataFrame(b)
```

```python
b
```

```python
c = pd.DataFrame(c)
c
```

```python
df = pd.concat([a,b,c])
df = df.reset_index(drop=True)
# combined
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
y_pred=kmeans.labels_
```

```python
plt.scatter(df[0],df[1],c=y_pred)
plt.title('KMeans')
```

```python
# the above graph shows the data for 3 cluster by kmean algo.
```

# Optimal Values DBSCAN

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbours=NearestNeighbors(n_neighbors=5)
neigh_fit = neighbours.fit(df)
distance, index=neigh_fit.kneighbors(df)
```

```python
distance = np.sort(distance,axis=0)
distances=distance[:,1]
plt.plot(distances)
plt.axhline(y=25,color='r')
plt.grid()
#plt.yticks(np.arange(0,130,10))
plt.ylim()
```

```python
# epsilon value found to be 25
```

```python
# optimum value for the min_sample
eps=25
for i in range(5,20):
    db=DBSCAN(eps=eps,min_samples=i)
    db.fit(df)
    labels=db.labels_
    
    new_labels=labels+2
    y=np.bincount(new_labels)
    z=np.nonzero(y)[0]
    cluster=list(zip(z,y[z]))
    print("min_samples:",i,"cluster:",cluster)
```

```python
db=DBSCAN(eps=eps,min_samples=6)
db.fit(df)
labels=db.labels_
```

```python
plt.scatter(df[0],df[1],c=labels)
```
